### matplotlib 패키지를 이용한 시각화

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris
```

```python
a = {'SepalLengthCm' : 'S_L', 'SepalWidthCm': 'S_W', 'PetalLengthCm': 'P_L',
    'PetalWidthCm' : 'P_W'}
iris2 = iris.rename(columns = a)
```

```python
iris2
```

```python
plt.hist(iris2['S_L'])
plt.xlabel('S_L')
plt.ylabel('freq')
plt.show()
```

```python
#산점도
plt.scatter(iris2['S_L'],iris2['S_W'])
plt.xlabel('S_L')
plt.ylabel('S_W')
plt.show()
```

```python
iris2[['S_L', 'S_W']].corr()
```

```python
pd.plotting.scatter_matrix(iris2[['S_L', 'S_W', 'P_L', 'P_W']])
plt.show()
```

```python
setosa = iris2[iris2.Species == 'Iris-setosa']
```

```python
setosa.head()
```

```python
plt.scatter(setosa['S_L'], setosa['S_W'])
plt.xlabel('S_L')
plt.ylabel('S_W')
plt.show()
```

```python
setosa[['S_L', 'S_W']].corr()
```

```python
versicolor = iris2[iris2.Species == 'Iris-versicolor']
```

```python
versicolor.head()
```

```python
plt.scatter(setosa['S_L'], versicolor['S_L'])
plt.xlabel("setosa['S_L']")
plt.ylabel("versicolor['S_L']")
plt.show()
```

```python
setosa_versicolor = pd.concat([setosa, versicolor],axis = 0)
```

```python
setosa_versicolor
```

```python
#문자열 항목 원그래프 작성
Species_nor = iris2['Species'].value_counts(ascending = False, normalize = True)
Species_nor
```

```python
Species_lab = Species_nor.index
Species_vals = Species_nor.values
print(Species_lab)
print(Species_vals)
```

```python
#레이블과 수치 따로 지정해서 그리기
plt.pie(Species_vals, labels = Species_lab)
plt.axis('equal')
plt.show()
```

```python
#분리안하고 그리기
plt.pie(Species_nor, labels = Species_lab)
plt.axis('equal')
plt.show()
```
