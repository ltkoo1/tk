```python
train_df = train.copy()
test_df = test.copy()
```

```python
train_df['MonthlyIncome'] = train_df['MonthlyIncome'].fillna(train_df.groupby('Occupation')['MonthlyIncome'].transform('mean'))
test_df['MonthlyIncome'] = test_df['MonthlyIncome'].fillna(test_df.groupby('Occupation')['MonthlyIncome'].transform('mean'))
```

```python
train_df['Age'] = train_df['Age'].fillna(train_df.groupby('Designation')['Age'].transform('mean'))
test_df['Age'] = test_df['Age'].fillna(test_df.groupby('Designation')['Age'].transform('mean'))
```

```python
from sklearn.impute import SimpleImputer
```

```python
fre_col = ['TypeofContact', 'DurationOfPitch', 'NumberOfFollowups', 'PreferredPropertyStar', 'NumberOfTrips', 'NumberOfChildrenVisiting']
imputer_fre = SimpleImputer(strategy = 'most_frequent')
imputer_fre.fit(train_df[fre_col])
train_df[fre_col] = imputer_fre.transform(train_df[fre_col])
test_df[fre_col] = imputer_fre.transform(test_df[fre_col])
```

```python
from pycaret.datasets import get_data
```

```python
from pycaret.classification import *
setup_clf = setup(data=train_df, target='ProdTaken', fold= 5, session_id=123) # fold 5 .
```

```python
top9 = compare_models(sort='Accuracy', n_select=9)
```

```python
tuned_top9 = [tune_model(i) for i in top9]
```

```python
blender_top9 = blend_models(estimator_list=tuned_top9)
```

```python
final_model = finalize_model(blender_top9)
prediction = predict_model(final_model, data=test_df)
```
